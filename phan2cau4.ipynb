```python
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import cm
```

```python
#đồ thị mặt yên ngựa
def mat_yen_ngua(x,y):
  x,y=np.meshgrid(x,y)
  z=(x/3)**2-(y/2)**2
  fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
  rosen_surf=ax.plot_surface(x,y,z,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
  ax.set_title('hình vẽ mặt yên ngựa') 
  ax.set_xlabel('trục hoành : x')
  ax.set_ylabel('trục tung : y')
  ax.set_zlabel('trục cao : z')
  plt.show()
```

```python
def main():
  x=np.linspace(-5,5)
  y=np.linspace(-5,5)
  mat_yen_ngua(x,y)
```

```python
if __name__=='__main__':
  main()
```

```python
#đồ thị mặt hyperbolic 1 tầng
def mat_hyperbolic(x,y):
  x,y=np.meshgrid(x,y)
# triển khai z ta có 2 nghiệm z1 và z2
  z1=2*((x/3)**2+(y/5)**2 - 1)**1/2
  z2=-2*((x/3)**2+(y/5)**2 - 1)**1/2
  fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
  rosen_surf=ax.plot_surface(x,y,z1,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  rosen_surf=ax.plot_surface(x,y,z2,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
  ax.set_xlabel('trục hoành')
  ax.set_ylabel('trục tung')
  ax.set_zlabel('trục cao')
  ax.set_title('hình vẽ mặt hyperbolic')
  plt.show()
```

```python
def main():
  x=np.linspace(-10,10)
  y=np.linspace(-10,10)
  mat_hyperbolic(x,y)
```

```python
if __name__=='__main__':
  main()
```

```python
import numpy as np
from numpy import*
import matplotlib.pyplot as plt 
from matplotlib import cm 
#đồ thị mặt yên ngựa
def mat_yen_ngua(x,y):
  x,y=np.meshgrid(x,y)
  z=(x/3)**2-(y/2)**2
  fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
  rosen_surf=ax.plot_surface(x,y,z,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
  ax.set_title('hình vẽ mặt yên ngựa') 
  plt.show()
#đồ thị mặt hyperbolic 1 tầng
def mat_hyperbolic(x,y):
  x,y=np.meshgrid(x,y)
# triển khai z ta có 2 nghiệm z1 và z2
  z1=2*((x/3)**2+(y/5)**2 - 1)**1/2
  z2=-2*((x/3)**2+(y/5)**2 - 1)**1/2
  fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
  rosen_surf=ax.plot_surface(x,y,z1,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  rosen_surf=ax.plot_surface(x,y,z2,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  fig.colorbar(rosen_surf, shrink=0.5,aspect=5)
  ax.set_title('hình vẽ mặt hyperbolic')
  plt.show()
# đồ thị mặt cầu
def mat_cau(a,b,c,d):
  k=np.linspace(0,2*pi,100)
  j=np.linspace(0,pi,100)
  x=(np.outer(np.cos(k),np.sin(j))*d)+a
  y=(np.outer(np.sin(k),np.sin(j))*d)+b
  z=(np.outer(np.ones(np.size(k)),np.cos(j))*d)+c
  fig,ax=plt.subplots(subplot_kw={'projection':'3d'})
  rosen_surf=ax.plot_surface(x,y,z,cmap=cm.gist_heat_r,linewidth=0,antialiased=False)
  fig.colorbar(rosen_surf, shrink=0.5,aspect=10)
  ax.set_title('hình vẽ mặt cầu')
  plt.show
```
